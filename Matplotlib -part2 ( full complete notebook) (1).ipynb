```python
'''
Python Matplotlib : Area Plot
Area plots are pretty much similar to the line plot. They are 
also known as 
stack plots. These plots can be used to track changes over time 
for two or more
related groups that make up one whole category. For example, 
lets compile the 
work done during a day into categories, say sleeping, eating, 
working and playing. Consider the below code: 
As we can see in the above image, we have time spent based on the
categories.
Therefore, area plot or stack plot is used to show trends over 
time, among 
different attributes
'''
```

```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,5]
playing = [8,5,7,8,10]
plt.plot([],[],color='r', label='Sleeping', linewidth=2)
plt.plot([],[],color='g', label='Eating', linewidth=2)
plt.plot([],[],color='y', label='Working', linewidth=2)
plt.plot([],[],color='c', label='Playing', linewidth=2)
plt.stackplot(days, playing,eating,working,sleeping)
plt.grid(True,color='b')  
plt.xlabel('x')
plt.ylabel('y')
plt.xticks([1,2,3,4,5])
#plt.yticks([])
plt.title('Stack Plot')
plt.legend(loc='best')
plt.savefig("e:\\abc.png")
plt.show()
```

```python
from matplotlib import pyplot as plt
from sklearn import datasets
import numpy as np
from sklearn.cluster import KMeans

iris1 = datasets.load_iris()
a = iris1.data[:,2:4]
b = iris1.target
plt.scatter(a[:,0], a[:,1], c=b)
kmeans_sk =KMeans(n_clusters=3, random_state=42)
kmeans_sk.fit(a)
clusters_sp= kmeans_sk.labels_
centroids_sp =kmeans_sk.cluster_centers_
plt.scatter(a[:,0], a[:,1], c=clusters_sp)
plt.scatter(centroids_sp[:,0],centroids_sp[:,1], marker='*', s = 200, c='#050505')
plt.show()
```
